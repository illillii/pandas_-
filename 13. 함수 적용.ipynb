## 데이터에 함수 적용 (apply)

```python
# 키 뒤에 cm 을 붙이는 역할
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm) # 키 데이터에 대해서 add_cm 함수를 호출한 결과 데이터를 반영
df
```

```python
def capitalize(lang):
    if pd.notnull(lang): # NaN 이 아닌지
        return lang.capitalize() # 첫 글자는 대문자로, 나머지는 소문자로
    return lang

df['SW특기'] = df['SW특기'].apply(capitalize)
df
```

# 함수 적용
함수 적용 요리조리

```python
import pandas as pd
df = pd.read_excel('school.xlsx', index_col='지원번호')
df
```

```python
df['성별'] = df['성별'] + '자'
df
```

```python
df['키'] = df['키'] + 'cm'
df
```

```python
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm)
df
```

```python
def capitalize(lang):
    if pd.notnull(lang):
        return lang.capitalize()
    return lang

df['SW특기'] = df['SW특기'].apply(capitalize)
df
```
